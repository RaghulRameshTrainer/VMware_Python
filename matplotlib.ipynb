```python
import matplotlib.pyplot as plt
```

```python
plt.plot([5,8,10],[6,15,7],linewidth=3,label="2021-profit",color="g")
plt.plot([6,7,11],[12,16,6],linewidth=3,label="2022-profit",color="b")

plt.title("P&L")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,3],label="Example-One",width=0.5)
plt.bar([2,4,6,8,10],[8,6,2,5,6],label="Example-Two",color="g",width=0.5)
plt.legend()
plt.title("Bar Graph")
plt.xlabel("height")
plt.ylabel("Weight")
plt.show()
```
