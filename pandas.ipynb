# HANDLING THE NULL / MISSING DATA

```python
df.head()
```

```python
df['Company'].mode()
```

```python

df['Company'].replace(np.nan,'TCS',inplace=True)
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].mean()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
df['Company'].value_counts()
```

```python
df['Company'].replace({'Tata Consultancy Services':'TCS','Infosys Pvt Lmt':'Infosys','Congnizant':'CTS'},inplace=True)
```

```python
df['Company'].value_counts()
```

```python
df.head()
```

```python
df.rename(columns={'Salary':'Pay'},inplace=True)
```

```python
df.head()
```

```python
df=df.drop(['Country','Gender'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['Place'].isnull().sum()
```

```python
df['Place'].replace({'NaN':np.nan},inplace=True)
```

```python
df=df.dropna(subset=['Place','Pay'])
```

```python
df['ExperienceLevel']='Senior'
```

```python
df=df.drop(columns=["ExperienceLevel"])
```

```python
df.loc[df['Age']>35,'ExperienceLevel']='Senior'
```

```python
df['ExperienceLevel']=df['ExperienceLevel'].fillna("Junior")
```

```python
df.loc[df['Pay']>=5000,'Category']="High Paid"
```

```python
df['Category']=df['Category'].fillna("Less Paid")
```

```python
df['Revised_Pay']=df['Pay'].apply(lambda x:x*2)
```

```python
def revise_sal(x):
    return x*1.2
```

```python
df['New_Pay']=df['Pay'].apply(revise_sal)
```

```python
def concat_values(x,y):
    return x+"-"+y
```

```python
df["Info"]=df.apply(lambda x:concat_values(x['Company'],x['Place']),axis=1)
```

```python
df['New_Age']=df['Age']*2
```
